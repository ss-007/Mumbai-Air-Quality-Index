```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from selenium import webdriver
import os
```

```python
browser = webdriver.Firefox(executable_path='C:\Drivers\geckodriver-v0.27.0-win64\geckodriver.exe')
```

```python
temp_data ={}
for year in range(2013,2019):
    Avg_temp = []
    Max_temp = []
    Min_temp = []
    Atm_pressure = []
    Avg_relative_humidity = []
    Total_rainfall_snowmelt= []
    Avg_visibility= []
    Avg_wind_speed= []
    Max_sustained_wind_speed= []
    
    for month in range(1,13):
        
        if month<10:
            browser.get("http://en.tutiempo.net/climate/0{}-{}/ws-432950.html".format(month, year))

            html = browser.page_source
            soup = BeautifulSoup(html, "html.parser")
            table = soup.find_all("table")
            row = table[3].select("tr")
            

            if month == 2:
                j = 29
                for i in row[1:j]:
                    Avg_temp.append(i.find_all("td")[1].text.strip())
                    Max_temp.append(i.find_all("td")[2].text.strip())
                    Min_temp.append(i.find_all("td")[3].text.strip())
                    Atm_pressure.append(i.find_all("td")[4].text.strip())
                    Avg_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Avg_visibility.append(i.find_all("td")[7].text.strip())
                    Avg_wind_speed.append(i.find_all("td")[8].text.strip())
                    Max_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                    
            elif month == 4 or month == 6 or month == 9:
                for i in row[1:31]:
                    Avg_temp.append(i.find_all("td")[1].text.strip())
                    Max_temp.append(i.find_all("td")[2].text.strip())
                    Min_temp.append(i.find_all("td")[3].text.strip())
                    Atm_pressure.append(i.find_all("td")[4].text.strip())
                    Avg_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Avg_visibility.append(i.find_all("td")[7].text.strip())
                    Avg_wind_speed.append(i.find_all("td")[8].text.strip())
                    Max_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                
                
            else:
                for i in row[1:32]:
                    Avg_temp.append(i.find_all("td")[1].text.strip())
                    Max_temp.append(i.find_all("td")[2].text.strip())
                    Min_temp.append(i.find_all("td")[3].text.strip())
                    Atm_pressure.append(i.find_all("td")[4].text.strip())
                    Avg_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Avg_visibility.append(i.find_all("td")[7].text.strip())
                    Avg_wind_speed.append(i.find_all("td")[8].text.strip())
                    Max_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                

        else:
            browser.get("http://en.tutiempo.net/climate/{}-{}/ws-432950.html".format(month, year))
            html = browser.page_source
            soup = BeautifulSoup(html, "html.parser")
            table = soup.find_all("table")
            row = table[3].select("tr")
            
            if month == 11:
                 for i in row[1:31]:
                    Avg_temp.append(i.find_all("td")[1].text.strip())
                    Max_temp.append(i.find_all("td")[2].text.strip())
                    Min_temp.append(i.find_all("td")[3].text.strip())
                    Atm_pressure.append(i.find_all("td")[4].text.strip())
                    Avg_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Avg_visibility.append(i.find_all("td")[7].text.strip())
                    Avg_wind_speed.append(i.find_all("td")[8].text.strip())
                    Max_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                    
                    
                    
            else:
                for i in row[1:32]:
                    Avg_temp.append(i.find_all("td")[1].text.strip())
                    Max_temp.append(i.find_all("td")[2].text.strip())
                    Min_temp.append(i.find_all("td")[3].text.strip())
                    Atm_pressure.append(i.find_all("td")[4].text.strip())
                    Avg_relative_humidity.append(i.find_all("td")[5].text.strip())
                    Total_rainfall_snowmelt.append(i.find_all("td")[6].text.strip())
                    Avg_visibility.append(i.find_all("td")[7].text.strip())
                    Avg_wind_speed.append(i.find_all("td")[8].text.strip())
                    Max_sustained_wind_speed.append(i.find_all("td")[9].text.strip())
                    
    
    data = pd.DataFrame({"Avg_temp":Avg_temp,
                        "Max_temp": Max_temp, 
                        "Min_temp" : Min_temp,
                        "Atm_pressure":Atm_pressure, 
                        "Avg_relative_humidity(%)": Avg_relative_humidity,
                        "Total_rainfall_snowmelt(mm)":Total_rainfall_snowmelt, 
                        "Avg_visibility": Avg_visibility,
                        "Avg_wind_speed": Avg_wind_speed, 
                        "Max_sustained_wind_speed": Max_sustained_wind_speed})
    
    temp_data[year] = data
```
